```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
titanic_df.info()
```

```python
titanic_df['sex'] = titanic_df['sex'].astype('category')
titanic_df['embarked'] = titanic_df['embarked'].astype('category')
titanic_df['who'] = titanic_df['who'].astype('category')
titanic_df['embark_town'] = titanic_df['embark_town'].astype('category')
titanic_df['alive'] = titanic_df['alive'].map({'yes': True, 'no': False})

titanic_df.info()
```

```python
titanic_df.describe(include='all').round(4)
```

A single column should be created indicating the number of relatives for each passenger on board, instead of:
- number of siblings or spouses on board;
- number of parents or children on board;
Boolean attribute: column 'alone'=True if the passenger was traveling alone (no relatives on board).

```python
titanic_df['relatives'] = titanic_df['sibsp'] + titanic_df['parch']
titanic_df.loc[titanic_df['alone'] == True, 'relatives'] = 0
titanic_df = titanic_df.drop(['sibsp', 'parch','alone'], axis=1)
titanic_df
```

```python
empty_values = titanic_df['relatives'].isna().sum()
print(empty_values)
```

To check how often a certain number of relatives appear in the new column.

```python
titanic_df_frq = titanic_df.groupby('relatives')['relatives'].agg(['count'])

titanic_df_frq
```

Using a loop, replace the number of relatives exceeding the number 5 with the value "above 5".

```python
titanic_df['relatives_category'] = titanic_df['relatives'].astype(object)

for i in titanic_df.index:
  if titanic_df.loc[i,'relatives'] > 5:
    titanic_df.loc[i, 'relatives_category'] = 'above 5'
#titanic_df.head(20)

titanic_df_relatives = titanic_df.groupby('relatives_category')['relatives_category'].agg(['count'])
titanic_df_relatives.head(20)
```

- count the total number of passengers in each of the cities of departure,
- find the number of passengers with more than 5 relatives in each of the cities of departure.

```python
titanic_df_stat = titanic_df.groupby('embark_town', observed=True)['embark_town'].count()
#print(titanic_df_stat)
titanic_df_pas_town = titanic_df[titanic_df['relatives_category'] == 'above 5'].groupby('embark_town', observed=True)['relatives_category'].count()
#print(titanic_df_pas_town)
result = (titanic_df_pas_town / titanic_df_stat * 100).round(2)
result = result.fillna(0)
result.name = 'percentage'
result
```

Fill in missing age values ​​with the median.

```python
median_age = titanic_df['age'].median()
#titanic_df['age'].fillna(median_age).info()
titanic_df['age'].fillna(median_age)
```

Create a new column where age is represented by a category instead of a number.

```python
def get_age_category(age):
  if pd.isna(age):
    return 'unknown'
  elif age < 14:
    return 'child'
  elif age < 35:
    return 'young'
  elif age < 59:
    return 'adult'
  else:
    return 'senior'

titanic_df['age_category'] = titanic_df['age'].apply(get_age_category)
titanic_df
```

Check which age group had the highest mortality rate.
Then count the total number of passengers and the number of those who did not survive.
Then calculate the relative rates for each category.

```python
dead_in_cat = titanic_df[titanic_df['alive'] == False].groupby('age_category')['alive'].count()
total_in_cat = titanic_df.groupby('age_category')['alive'].count()
rate_in_cat = (dead_in_cat / total_in_cat * 100).round(2)

print('\nDead in cat:', dead_in_cat) #total number of those who did not survive for each category
print('\nTotal in cat:', total_in_cat) #total number of passengers for each category
print('\nRate in %:',rate_in_cat)
print('\n', rate_in_cat.idxmax())
```

Detailed mortality statistics by age category, ticket class, cabin level, and number of relatives.

```python
dead_in_cat_age = titanic_df[titanic_df['alive'] == False].groupby('age_category')['alive'].count()
total_in_cat_age = titanic_df.groupby('age_category')['alive'].count()
rate_in_cat_age = (dead_in_cat_age / total_in_cat_age * 100).round(2)
print('Age category:', rate_in_cat_age)
print('\n', rate_in_cat_age[rate_in_cat_age == rate_in_cat_age.max()])

dead_in_cat_class = titanic_df[titanic_df['alive'] == False].groupby('class', observed=True)['alive'].count()
total_in_cat_class = titanic_df.groupby('class', observed=True)['alive'].count()
rate_in_cat_class = (dead_in_cat_class / total_in_cat_class * 100).round(2)
print('\nClass category:', rate_in_cat_class)
print('\n', rate_in_cat_class[rate_in_cat_class == rate_in_cat_class.max()])

dead_in_cat_deck = titanic_df[titanic_df['alive'] == False].groupby('deck', observed=True)['alive'].count()
total_in_cat_deck = titanic_df.groupby('deck', observed=True)['alive'].count()
rate_in_cat_deck = (dead_in_cat_deck / total_in_cat_deck * 100).round(2)
print('\nDeck category:', rate_in_cat_deck)
print('\n', rate_in_cat_deck[rate_in_cat_deck == rate_in_cat_deck.max()])

dead_in_cat_relatives = titanic_df[titanic_df['alive'] == False].groupby('relatives', observed=True)['alive'].count()
total_in_cat_relatives = titanic_df.groupby('relatives', observed=True)['alive'].count()
rate_in_cat_relatives = (dead_in_cat_relatives / total_in_cat_relatives * 100).round(2)
print('\nRelatives category:', rate_in_cat_relatives)
print('\n', rate_in_cat_relatives[rate_in_cat_relatives == rate_in_cat_relatives.max()])
```

The highest mortality rate was observed among passengers in the senior age group, with a mortality rate of 75.00%.

Among ticket classes, third-class passengers experienced the highest mortality rate at 75.76%.

Passengers traveling with 7 or 10 relatives had a 100% mortality rate, meaning no one in these groups survived.

Regarding cabin levels, deck G was associated with the highest mortality rate at 50.00%, making it the riskiest cabin deck.

Overall, ticket class and age appear to have had the greatest impact on passenger mortality.
